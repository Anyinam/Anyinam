```python
# Execute the INSERT query
cursor.execute("INSERT INTO anime_Producer (title, Producer) "
               "SELECT title, SUBSTRING_INDEX(SUBSTRING_INDEX(Producers, ', ', numbers.n), ', ', -1) Producers "
               "FROM ("
               "    SELECT 1 n UNION ALL SELECT 2 UNION ALL SELECT 3 UNION ALL SELECT 4 UNION ALL SELECT 5"
               ") numbers INNER JOIN Producers "
               "ON CHAR_LENGTH(Producers) - CHAR_LENGTH(REPLACE(Producers, ', ', '')) >= (numbers.n - 1) * 2;")

# Commit the changes
conn.commit()
```
