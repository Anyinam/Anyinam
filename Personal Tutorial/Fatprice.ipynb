```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("second.csv")
df
```

```python
df.rename(columns={'Russian GDP': 'Rgdp'}, inplace=True)
df.rename(columns={'Russia Oil Proved Reserve(in billions barrels)':"Roreserve"}, inplace=True)
df.rename(columns={'Global oil consumption(mb/d)': 'Gcon'}, inplace=True)
```

```python
plt.plot(df["Year"], df["Oprice"])
plt.xlabel("YEAR")
plt.ylabel("WTI(OIL PRICE)")
plt.title("WTI TREND")
plt.show()
```

```python
plt.scatter(df["Rgdp"], df["Oprice"])
plt.xlabel("Russia GDP")
plt.ylabel("Oil Price")
plt.title("Distribution of Oil price to Russia GDP")
plt.show()
```

```python
plt.scatter(df["Roreserve"], df["Oprice"])
plt.xlabel("Russia Oil Proved Reserve")
plt.ylabel("Oil Price")
plt.title("Distribution of Oil price to Russia Oil Proved Reserve")
plt.show()
```

```python
plt.scatter(df["Gcon"], df["Oprice"])
plt.xlabel("Global Oil consumption")
plt.ylabel("Oil Price")
plt.title("Distribution of Oil price to Global Oil consumption")
plt.show()
```
